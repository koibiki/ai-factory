```python
def k_fold_validation(X, Y):
    print('开始CV 5折训练...')
    kf = KFold(len(X), n_folds=10, shuffle=True, random_state=33)
    kf_X_trains = []
    kf_y_trains =[]
    kf_X_valids= []
    kf_y_valids = []
    for i, (train_index, test_index) in enumerate(kf):
        print('第{}次训练...'.format(i))
        kf_X_train = X.iloc[train_index]
        kf_y_train = Y.iloc[train_index]
        kf_X_valid = X.iloc[test_index]
        kf_y_valid = Y.iloc[test_index]
        kf_X_trains.append(kf_X_train)
        kf_y_trains.append(kf_y_train)
        kf_X_valids.append(kf_X_valid)
        kf_y_valids.append(kf_y_valid)
    return  kf_X_trains,kf_y_trains,kf_X_valids,kf_y_valids
```

```python
lgm = LightGBM()
kf_X_trains,kf_y_trains,kf_X_valids,kf_y_valid = k_fold_validation(train_X, train_Y)
```
